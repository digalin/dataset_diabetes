```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, MaxAbsScaler, QuantileTransformer
from joblib import dump
```

```python
df_data = pd.read_csv("../results/2_check_outliers.csv")
df_data
```

```python
df_col_not_work = df_data[["is_isolated", "is_lof", "Outcome"]]
df_col_to_work = df_data.drop(columns=["is_isolated", "is_lof", "Outcome"])
df_col_to_work
```

```python
robust_scaler = RobustScaler()
robust_scaler.fit(df_col_to_work.values)
df_robust_scaled = robust_scaler.transform(df_col_to_work.values)
```

```python
df_robust_scaled
```

```python
standard_scaler = StandardScaler()
standard_scaler.fit(df_robust_scaled)
df_standard_scaled = standard_scaler.transform(df_robust_scaled)
df_standard_scaled
```

```python
# Diego Galindo, Francisco Mercado

df_scaled = pd.DataFrame(data=df_standard_scaled, columns=df_col_to_work.columns)
df_final = pd.concat([df_scaled, df_col_not_work.reset_index(drop=True)], axis=1)
df_final
```

```python
df_final.to_csv("../results/3_standardization_final.csv", index=False)
```

```python
dump(standard_scaler, "../results/3_standard_scaler_instance.joblib")
```
